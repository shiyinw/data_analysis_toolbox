```r
iris2 = tbl_df(iris)
head(iris2)
```

```r
junk = mutate(iris2, Sepal.Size = Sepal.Length + Sepal.Width)
head(junk)
# 生成新的数据维度
```

```r
summarize(group_by(iris, Species), Mean.Petal.Size = mean(Petal.Width+Petal.Length), Mean.Sepal.Size = mean(Sepal.Width+Sepal.Length))
```

```r
k = group_by(iris, Species)
k["setosa"]
iris[iris$Species == "setosa"]$Petal.Width
```
